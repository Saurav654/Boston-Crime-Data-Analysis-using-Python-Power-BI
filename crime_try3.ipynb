```python
for street, coord_list in street_to_coords.items():
    for lat, lon in coord_list:
        df.loc[
            (df["STREET"].isna()) &
            (df["Lat"] == lat) &
            (df["Long"] == lon),
            "STREET"
        ] = street
```

```python
for street, rep_list in SR.items():
    if len(rep_list) == 1: 
        df.loc[
            (df["REPORTING_AREA"].isna()) &
            (df["STREET"] == street),
            "REPORTING_AREA"
        ] = rep_list[0]
```

```python
RD=df[(df["DISTRICT"].notna()) & (df["REPORTING_AREA"].notna()) ].groupby(["REPORTING_AREA"])["DISTRICT"].unique().to_dict()
for rep,dist_list in RD.items():
    if len(dist_list) == 1: 
        df.loc[
            (df["DISTRICT"].isna()) &
            (df["REPORTING_AREA"] == rep),
            "DISTRICT"
        ] = dist_list[0]
```

```python
df.isna().sum()
```

```python
dist_ra_coords = (
    df[
        df["DISTRICT"].notna() &
        df["REPORTING_AREA"].notna() &
        df["Lat"].notna() &
        df["Long"].notna()
    ]
    .groupby(["DISTRICT", "REPORTING_AREA"])[["Lat", "Long"]]
    .mean()
)
```

```python
dist_ra_coords
```

```python
dist_to_ra = (
    df[df["DISTRICT"].notna() & df["REPORTING_AREA"].notna()]
    .groupby("DISTRICT")["REPORTING_AREA"]
    .unique()
    .to_dict()
)
```

```python
street_to_ra = (
    df[df["STREET"].notna() & df["REPORTING_AREA"].notna()]
    .groupby("STREET")["REPORTING_AREA"]
    .unique()
)

street_to_ra = street_to_ra[street_to_ra.apply(len) == 1].to_dict()
```

```python
street_to_ra
```

```python
mask = df["REPORTING_AREA"].isna() & df["DISTRICT"].notna()

for idx in df[mask].index:
    district = df.at[idx, "DISTRICT"]
    lat = df.at[idx, "Lat"]
    lon = df.at[idx, "Long"]
    street = df.at[idx, "STREET"]

    # 1 LatLong based
    if pd.notna(lat) and pd.notna(lon) and district in dist_ra_coords.index.get_level_values(0):
        sub = dist_ra_coords.loc[district]
        distances = ((sub["Lat"] - lat) ** 2 + (sub["Long"] - lon) ** 2) ** 0.5
        df.at[idx, "REPORTING_AREA"] = distances.idxmin()

    # 2 Street-based (safe only)
    elif pd.notna(street) and street in street_to_ra:
        df.at[idx, "REPORTING_AREA"] = street_to_ra[street][0]

    # 3 Random fallback
    elif district in dist_to_ra:
        df.at[idx, "REPORTING_AREA"] = np.random.choice(dist_to_ra[district])
```

```python
df.isna().sum()
```

```python


ra_street_coords = (
    df[
        df["REPORTING_AREA"].notna() &
        df["STREET"].notna() &
        df["Lat"].notna() &
        df["Long"].notna()
    ]
    .groupby(["REPORTING_AREA", "STREET"])[["Lat", "Long"]]
    .mean()
)
```

```python
ra_street_coords
```

```python
ra_to_street = (
    df[df["REPORTING_AREA"].notna() & df["STREET"].notna()]
    .groupby("REPORTING_AREA")["STREET"]
    .unique()
    .to_dict()
)
```

```python
ra_to_street
```

```python
mask = df["STREET"].isna() & df["REPORTING_AREA"].notna()

for idx in df[mask].index:
    ra = df.at[idx, "REPORTING_AREA"]
    lat = df.at[idx, "Lat"]
    lon = df.at[idx, "Long"]

    # 1 LatLong based fill
    if (
        pd.notna(lat) and
        pd.notna(lon) and
        ra in ra_street_coords.index.get_level_values(0)
    ):
        sub = ra_street_coords.loc[ra]

        distances = np.sqrt(
            (sub["Lat"] - lat) ** 2 +
            (sub["Long"] - lon) ** 2
        )

        df.at[idx, "STREET"] = distances.idxmin()

    # 2 Random fallback
    elif ra in ra_to_street:
        df.at[idx, "STREET"] = np.random.choice(ra_to_street[ra])
```

```python
SLLR = {
    i: df[
        (df["STREET"] == i) &
        df["Lat"].notna() &
        df["Long"].notna()
    ][["Lat", "Long"]]
    .drop_duplicates()
    .values
    .tolist()
    for i in df[df["STREET"].notna()]["STREET"].unique().tolist()
}
```

```python

for street, coords in SLLR.items():
    if len(coords) > 0:
        lat, lon = coords[np.random.randint(len(coords))]

        df.loc[
            (df["STREET"] == street) &
            df["Lat"].isna() &
            df["Long"].isna(),
            ["Lat", "Long"]
        ] = [lat, lon]
```
